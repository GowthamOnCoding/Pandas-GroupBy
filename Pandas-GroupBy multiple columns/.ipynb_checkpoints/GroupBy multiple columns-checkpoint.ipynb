# GroupBy multiple columns

```python
import pandas as pd
df=pd.read_csv("fortune1000.csv",index_col="Rank")
df.head()
```

```python
#Lets group by Sector then Industry
gb_obj=df.groupby(["Sector","Industry"])
gb_obj
```

```python
gb_obj.size()
```

```python
gb_obj["Profits"].sum()
```

```python
gb_obj[["Employees","Profits"]].mean()
```

# The .agg() method

```python
sector=df.groupby("Sector")
```

```python
sector.agg({"Employees":"sum",
           "Profits":"mean",
           "Revenue":"min"})
```

```python
sector.agg({"Profits":["sum","min","mean"],
          "Revenue":"sum"})
```

```python
sector.agg(["mean","sum","max"])
```

# Iterating group values

```python
df=pd.read_csv("fortune1000.csv",index_col="Rank")
df.head(3)
```

```python
industry=df.groupby("Industry")
```

```python
#Lets find out the highest revenue of each industry
#Lets create empty dataframe to append our resulting dataframe
df2=pd.DataFrame(columns=df.columns)
df2
```

```python
#Iterate each group using for loop
for group,data in industry:
   
    df2=df2.append(data.nlargest(1,"Profits"))
```
